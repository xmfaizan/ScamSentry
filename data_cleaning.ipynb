```python
import pandas as pd
```

```python
df = pd.read_csv('fake reviews dataset.csv')
df.head()
```

```python
df.describe()
```

```python
df.value_counts().sum()
```

```python
df.isnull().sum()
```

```python
df.drop(['category', 'rating'], axis=1, inplace=True)
df.head()
```

```python
df = df.rename(columns={'label': 'labels', 'text_': 'texts'})
df.head()
```

```python
df['labels'] = df['labels'].replace('CG', 0)
df['labels'] = df['labels'].replace('OR', 1)

cg_count = df['labels'].value_counts()[0]
or_count = df['labels'].value_counts()[1]

print(f"Number of CG values: {cg_count}")
print(f"Number of OR values: {or_count}")
```

*   CG - Computer Generated
*   OR - Original Review



```python
df.to_csv('cleaned.csv', index=False)
```
